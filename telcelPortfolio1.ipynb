```python
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
# SQL Alchemy
from sqlalchemy import create_engine
# PyMySQL 
from selenium.common.exceptions import TimeoutException, WebDriverException
import time
from sqlalchemy import create_engine

```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
selenium_driver = browser.driver
```

```python
featured_image_url = 'https://www.telcel.com/personas/equipos/telefonos-y-smartphones'
browser.visit(featured_image_url)
```

```python
list_promotion =[]
list_handset_OEM = []
list_handset_model = []
list_handset_list_price = []
list_handset_sale_price =[]
list_handset_ids =[]
contador = 0

while True:
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    handsets = soup.find_all('div', class_='telcel-mosaico-equipos-detalle')
    
    for handset in handsets:
        list_handset_ids.append(contador)
        contador = contador + 1
        list_handset_OEM.append(handset.find('p', class_="telcel-mosaico-equipos-marca ng-binding").text)
        list_handset_model.append(handset.find('p', class_="telcel-mosaico-equipos-nombre-equipo ng-binding").text)
        list_handset_list_price.append(handset.find('span', class_= "ng-binding").text)
        list_handset_sale_price.append(handset.find('p', class_= "telcel-mosaico-equipos-precio").text)
    try:
        selenium_driver.execute_script("window.scrollTo(0, 1500);")
    except WebDriverException as e:
        print ("WebdriverException")       
    try:    
        boton = selenium_driver.find_element_by_css_selector("a[aria-label='Página siguiente']").click()
    except WebDriverException as e:
        print("Selenium Exception: {0} Message: {1}".format("my message", str(e)))
        break
    contador = contador + 1





 
```

```python
handset_list = {
         'ID': list_handset_ids,
         'OEM': list_handset_OEM,
         'Handset': list_handset_model,
         'List Price': list_handset_list_price,
         'Sale Price': list_handset_sale_price}
    
handsets_telcel_df= pd.DataFrame(handset_list)
handsets_telcel_df.set_index('ID', inplace = True)

handsets_telcel_df.head()
```

```python
handsets_telcel_df['OEM'] = handsets_telcel_df["OEM"].str.rpartition("|")
handsets_telcel_df
```

```python
handsets_telcel_df['List Price2'] = handsets_telcel_df['List Price']

handsets_telcel_df
```
